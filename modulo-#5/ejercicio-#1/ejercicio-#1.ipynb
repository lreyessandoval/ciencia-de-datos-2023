Aprendizaje supervisado

El Aprendizaje supervisado es un tipo de aprendizaje de máquina en el que se 
proporcionan datos de entrenamiento etiquetados. El algoritmo de aprendizaje
supervisado aprende de estos datos etiquetados para predecir la etiqueta de
datos nuevos o sin etiquetar. El aprendizaje supervisado se puede dividir en
dos categorías: clasificación y regresión. La clasificación es cuando la
variable de destino es categórica, mientras que la regresión es cuando la
variable de destino es numérica.


Aprendizaje no supervisado

 El aprendizaje no supervisado es un tipo de aprendizaje automático en el que
 no se proporcionan datos de entrenamiento etiquetados. El algoritmo de
 aprendizaje no supervisado aprende de los datos sin etiquetar para descubrir
 patrones en los datos. Debido a la falta de datos etiquetados, no hay
 retroalimentación correcta y, por lo tanto, no hay forma de evaluar la
 precisión del modelo. El aprendizaje no supervisado se puede dividir en dos
 categorías: agrupación y asociación. La agrupación es cuando el objetivo es
 descubrir grupos de puntos de datos similares en un conjunto de datos. La
 asociación es cuando el objetivo es descubrir reglas que describan grandes
 porciones de su conjunto de datos.

Aprendizaje por refuerzo

 El aprendizaje por refuerzo es un tipo de aprendizaje automático en el que el
 algoritmo de aprendizaje recibe retroalimentación dinámica de su entorno en
 forma de recompensas para tomar decisiones óptimas o alcanzar un objetivo.
 Cada decisión que toma el algoritmo de aprendizaje tiene un impacto en el
 entorno y la retroalimentación que recibe. El objetivo del algoritmo de
 aprendizaje es encontrar la mejor política, una estrategia que maximice la
 recompensa a largo plazo. El aprendizaje por refuerzo se puede dividir en dos
 categorías: aprendizaje con y sin modelo. El aprendizaje con modelo es cuando
 el algoritmo de aprendizaje conoce el modelo del entorno. El aprendizaje sin
 modelo es cuando el algoritmo de aprendizaje no conoce el modelo del entorno.

Algoritmos de regresión

 Los algoritmos de regresión son algoritmos de aprendizaje automático que se
 utilizan para predecir una variable numérica. La variable de destino se
 representa como una línea de mejor ajuste (regresión lineal) o una curva
 (regresión polinómica) para predecir el valor de la variable de destino dado
 el valor de la variable independiente. Los algoritmos de regresión se pueden
 dividir en dos categorías: regresión lineal y regresión no lineal. La
 regresión lineal es cuando la variable de destino se puede representar como
 una línea de mejor ajuste. La regresión no lineal es cuando la variable de
 destino se puede representar como una curva.

 Algoritmos de agrupación

Los algoritmos de agrupación son algoritmos de aprendizaje automático que se
utilizan para agrupar puntos de datos similares en un conjunto de datos. Los
algoritmos de agrupación se pueden dividir en dos categorías: agrupación
jerárquica y agrupación no jerárquica. La agrupación jerárquica es cuando los
puntos de datos se agrupan en función de la similitud entre ellos. La
agrupación no jerárquica es cuando los puntos de datos se agrupan en función
de la similitud entre ellos y el número de grupos se determina de antemano.

Algoritmos de Aprendizaje Profundo

Los algoritmos de aprendizaje profundo son algoritmos de aprendizaje
automático que se utilizan para extraer características de los datos. Los
algoritmos de aprendizaje profundo se pueden dividir en dos categorías:
aprendizaje profundo supervisado y aprendizaje profundo no supervisado. El
aprendizaje profundo supervisado es cuando el algoritmo de aprendizaje
profundo se entrena con datos etiquetados. El aprendizaje profundo no
supervisado es cuando el algoritmo de aprendizaje profundo se entrena con
datos sin etiquetar.

Qué es un algoritmo

Un algoritmo es un conjunto de instrucciones que se utilizan para resolver
un problema o realizar una tarea. Los algoritmos se pueden escribir en
cualquier lenguaje de programación, incluidos Python, C, Java, etc. Los
algoritmos se pueden diseñar utilizando diagramas de flujo o pseudocódigo.

Qué tipos de aprendizaje existen.

Los tipos de aprendizaje son aprendizaje supervisado, aprendizaje no
supervisado y aprendizaje por refuerzo.
