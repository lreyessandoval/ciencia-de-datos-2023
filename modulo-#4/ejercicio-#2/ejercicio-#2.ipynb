```python
# Realizar cálculos de probabilidades seleccionando la distribución de probabilidades para resolver un problema.
# variable Aleatoria en Distribución normal
# Distribución normal

# Si X es una variable aleatoria de una distribución normal con media μ y desviación estándar.
# hallar: P(μ - 3σ <= X <= μ  + 3σ)

# P(μ - 3σ <= X <= μ  + 3σ) = P(X <= μ  + 3σ) - P(X <= μ - 3σ)
# P(X <= μ  + 3σ) = 1 - P(X > μ  + 3σ)
# P(X <= μ - 3σ) = 1 - P(X > μ - 3σ)

# P(X <= μ  + 3σ) = 1 - P(Z > (μ  + 3σ - μ)/σ)
# P(X <= μ - 3σ) = 1 - P(Z > (μ - 3σ - μ)/σ)

# P(X <= μ  + 3σ) = 1 - P(Z > 3)
# P(X <= μ - 3σ) = 1 - P(Z > -3)

# P(X <= μ  + 3σ) = 1 - 0.0013
# P(X <= μ - 3σ) = 1 - 0.9987

# P(X <= μ  + 3σ) = 0.9987
# P(X <= μ - 3σ) = 0.0013

# P(μ - 3σ <= X <= μ  + 3σ) = 0.9987 - 0.0013
# P(μ - 3σ <= X <= μ  + 3σ) = 0.9974

import numpy as np
import matplotlib.pyplot as plt

mu = 0  # media
sigma = 1  # desviación estándar

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = np.exp(-(x - mu)**2 / (2*sigma**2)) / (sigma * np.sqrt(2*np.pi))

plt.plot(x, y)
plt.title('Distribución normal')
plt.xlabel('Valores de x')
plt.ylabel('Densidad de probabilidad')


plt.axvline(x=-0.9974, color='green', linestyle='--')
plt.axvline(x=0.9974, color='green', linestyle='--')

plt.show()
```

```python
# Caso 2: Distribución normal, media y desviación típica.
# En una distribución normal de media 4 y desviación típica 2, calcular el valor de a para que:
# P(4 - a <= x <= 4 + a) = 0.5934

# Para resolver este problema, debemos utilizar las propiedades de la distribución normal y encontrar el valor de a para que P(4-a ≤ x ≤ 4+a) = 0.5934.
# Primero, podemos estandarizar la variable aleatoria x utilizando la fórmula de estandarización:

# Z = (X - μ) / σ

# Donde X es la variable aleatoria normal que sigue la distribución normal de media μ y desviación típica σ, y Z es la variable aleatoria normal estándar.
# En este caso, X tiene una media de 4 y una desviación típica de 2, por lo que la estandarización de X es:
# Z = (X - 4) / 2

# La probabilidad que queremos encontrar es:
# P(4-a ≤ x ≤ 4+a) = P((4-a - 4) / 2 ≤ (x - 4) / 2 ≤ (4+a - 4) / 2) = P(-a / 2 ≤ Z ≤ a / 2)
# Ahora, necesitamos encontrar el valor de a para que P(-a/2 ≤ Z ≤ a/2) = 0.5934. Podemos utilizar una tabla de distribución normal estándar para encontrar el valor correspondiente de la probabilidad acumulada de Z a ambos lados:
# P(Z ≤ a/2) - P(Z ≤ -a/2) = 0.5934
# De la tabla, sabemos que P(Z ≤ -a/2) = 1 - P(Z ≤ a/2). Entonces, podemos despejar el valor de P(Z ≤ a/2):
# P(Z ≤ a/2) = (1 + 0.5934) / 2 = 0.7967
# Usando la tabla de distribución normal estándar de nuevo, podemos encontrar el valor correspondiente de Z que produce esta probabilidad acumulada:
# a/2 = 0.833
# Despejando a, tenemos:
# a = 1.666
# Por lo tanto, para que P(4-a ≤ x ≤ 4+a) = 0.5934, el valor de a debe ser aproximadamente 1.666.



import numpy as np
import matplotlib.pyplot as plt

# Definimos los parámetros de la distribución normal
mu = 4  # media
sigma = 2  # desviación típica

# Generamos una muestra de valores aleatorios de la distribución normal
num_samples = 1000
samples = np.random.normal(mu, sigma, num_samples)

# Graficamos un histograma de los valores aleatorios generados
plt.hist(samples, bins=30, density=True, alpha=0.5, color='blue')

# Generamos una serie de valores de x para graficar la función de densidad de probabilidad
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))

# Graficamos la función de densidad de probabilidad
plt.plot(x, y, color='red')

# Ajustamos los ejes y agregamos etiquetas y título
plt.xlim(mu - 4*sigma, mu + 4*sigma)
plt.xlabel('Valores de x')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribución normal con media 4 y desviación típica 2')

a=1.666
plt.axvline(x=4-a, color='green', linestyle='--')
plt.axvline(x=4+a, color='green', linestyle='--')

# Mostramos el gráfico
plt.show()


```

```python
# Caso 3: Distribución Normal
# Dada una variable aleatoria continua Z, con distribución normal estándar, es decir, N(0;1),
# encuentre las siguientes probabilidades, usando la tabla.

import pandas as pd
from scipy.stats import norm

# Creamos un DataFrame con los valores de Z y las probabilidades acumuladas
start = 0
end = 3
step = 1.33
z_values = pd.DataFrame({'Z': np.arange(start, end + step, step)}) # valores de Z
z_values['Probabilidad'] = norm.cdf(z_values['Z']) # probabilidades acumuladas

# a) P(0 ≤ Z ≤ 1.25)
prob_a = norm.cdf(1.25) - norm.cdf(0) # P(0 ≤ Z ≤ 1.25) = P(Z ≤ 1.25) - P(Z ≤ 0)
print(f"a) P(0 ≤ Z ≤ 1.25) = {prob_a:.4f}") 

# b) P(Z ≥ 1.25)
prob_b = 1 - norm.cdf(1.25)
print(f"b) P(Z ≥ 1.25) = {prob_b:.4f}") 

# c) P(Z ≤ -1.25)
prob_c = norm.cdf(-1.25)
print(f"c) P(Z ≤ -1.25) = {prob_c:.4f}") 

# d) P(0 ≤ Z ≤ 1.33)
prob_d = norm.cdf(1.33) - norm.cdf(0)
print(f"d) P(0 ≤ Z ≤ 1.33) = {prob_d:.4f}") 

# e) P(Z ≥ 1.33)
prob_e = 1 - norm.cdf(1.33)
print(f"e) P(Z ≥ 1.33) = {prob_e:.4f}") 

# f) P(-1.33 ≤ Z ≤ 0)
prob_f = norm.cdf(0) - norm.cdf(-1.33)
print(f"f) P(-1.33 ≤ Z ≤ 0) = {prob_f:.4f}") 

# Mostramos los valores de Z y las probabilidades acumuladas
print("\nValores de Z y probabilidades acumuladas:")
print(z_values)

#graficamos
import numpy as np
import matplotlib.pyplot as plt

# Definimos los parámetros de la distribución normal
mu = 0  # media
sigma = 1  # desviación típica

# Generamos una muestra de valores aleatorios de la distribución normal
num_samples = 1000
samples = np.random.normal(mu, sigma, num_samples)

# Graficamos un histograma de los valores aleatorios generados
plt.hist(samples, bins=30, density=True, alpha=0.5, color='blue')

# Generamos una serie de valores de x para graficar la función de densidad de probabilidad
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))

# Graficamos la función de densidad de probabilidad
plt.plot(x, y, color='red')

# Ajustamos los ejes y agregamos etiquetas y título
plt.xlim(mu - 4*sigma, mu + 4*sigma)
plt.xlabel('Valores de x')
plt.ylabel('Densidad de probabilidad')
plt.title('Distribución normal con media 0 y desviación típica 1')

plt.axvline(x=0, color='green', linestyle='--')
plt.axvline(x=1.25, color='green', linestyle='--')
plt.axvline(x=-1.25, color='green', linestyle='--')
plt.axvline(x=1.33, color='green', linestyle='--')
plt.axvline(x=-1.33, color='green', linestyle='--')

# Mostramos el gráfico
plt.show()
```

```python
# Caso 4: Distribución Binomial.

# Uno de los ítems de una prueba sorpresa consta de 12 preguntas de verdadero o falso.
# Suponiendo que los estudiantes no saben contestar a ninguna de ellas y, contestan al
# azar, hallar la probabilidad de obtener seis aciertos.

# Solución:
# Sea X la variable aleatoria que representa el número de aciertos en las 12 preguntas.
# Entonces, X sigue una distribución binomial con n = 12 y p = 0.5.
# La probabilidad de obtener seis aciertos es P(X = 6).

from scipy.stats import binom

# Definimos los parámetros de la distribución binomial
n = 12  # número de preguntas
p = 0.5  # probabilidad de acertar

# Calculamos la probabilidad de obtener 6 aciertos
prob = binom.pmf(6, n, p)
print(f"La probabilidad de obtener 6 aciertos es {prob:.4f}")
```
