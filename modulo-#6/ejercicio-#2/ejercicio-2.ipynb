```python
# Clasificación de Clientes según su personalidad.

# Luego de haber cerrado su cápsula de aprendizaje de clasificación, se le ha
# determinado que averigue los intereses que pueden tener los clientes de la
# empresa KEPPLER, esto es necesario, para que se entregue la información al
# departamento de marketing y puedan realizar campañas para fidelizar a sus
# clientes.

# En el archivo clientes.cvs encontrará el registro de cada uno de ellos,
# además el archivo contiene 9 categorías -actividades laborales- que son:

# Actor/actriz
# Cantante
# Modelo
# Tv, series
# Radio
# Tecnología
# Deportes
# Politica
# Escritor

```

```python
# 1. Realice un resumen estadístico de sus datos.

# carga de datos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('clientes.csv', sep=',')
df.head()

# resumen estadístico
df.describe()

```

```python
# 2. Entregue gráficos que acompañen los hallazgos encontrados en el punto 1.

# gráfico de barras
df['categoria'].value_counts().plot(kind='bar')
```

```python
# calcular correlacion excluyendo la columna 'usuario'
df.drop(['usuario'], axis=1, inplace=True)

corr = df.corr()
corr

# graficar correlacion
plt.figure(figsize=(20,20))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
# 3. Encuentre una relación o grafique las variables: op, ex y ag, además de
# las categorías.

# graficar op, ex y ag
sns.pairplot(df, hue='categoria', vars=['op', 'ex', 'ag'])
plt.show()
```

```python
# 4. Determine cuántos k o clúster son adecuados para esta consulta.

# importar librerias
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# warning
import warnings
warnings.filterwarnings('ignore')


# calcular el coeficiente de silueta para diferentes valores de k
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k).fit(df)
    print("Para k=%s, el coeficiente de silueta es %s" % (k, silhouette_score(df, kmeans.labels_)))

# graficar coeficiente de silueta
plt.plot(range(2, 10), [silhouette_score(df, KMeans(n_clusters=i).fit(df).labels_) for i in range(2, 10)], "bo-")
plt.title("Coeficiente de silueta vs número de clusters")
plt.xlabel("Número de clusters")
plt.ylabel("Coeficiente de silueta")
plt.show()


```

```python
# 5. Realice un modelo de k means con los cluster definidos en el punto 4.

# importar librerias
from sklearn.cluster import KMeans

# definir el modelo
kmeans = KMeans(n_clusters=3, random_state=0)

# entrenar el modelo
kmeans.fit(df)

# predecir
pred = kmeans.predict(df)

# agregar columna 'cluster' al dataframe
df['cluster'] = pred

# mostrar dataframe
df.head()


```

```python
# 6. Entregue una representación gráfica de los cluster.

# graficar cluster
sns.pairplot(df, hue='cluster', vars=['op', 'ex', 'ag'])
plt.show()
```
