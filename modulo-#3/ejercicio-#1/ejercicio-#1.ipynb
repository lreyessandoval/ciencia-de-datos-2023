```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


metros = [1,2,3,4,5,6,7,8]
niños = [2,6,10,5,10,3,2,2]

#a) Tabla de Frecuencia. Diagrama de Barras para las frecuencias absolutas, relativas y acumuladas.

df = pd.DataFrame({'Metros': metros, 'Niños': niños})
df['Frecuencia Absoluta'] = df.Niños.value_counts()
df['Frecuencia Relativa'] = df.Niños.value_counts(normalize=True)
df['Frecuencia Acumulada'] = df['Frecuencia Relativa'].cumsum()


# grafico de barras para frecuencia absoluta
df['Frecuencia Absoluta'].plot(kind='bar', title='Frecuencia Absoluta', xlabel='Metros', ylabel='Niños')
plt.show()
df['Frecuencia Relativa'].plot(kind='bar', title='Frecuencia Relativa', xlabel='Metros', ylabel='Niños')
plt.show()
df['Frecuencia Acumulada'].plot(kind='bar', title='Frecuencia Acumulada', xlabel='Metros', ylabel='Niños')
plt.show()

```

```python
#b. Mediana, media aritmética, moda y cuartiles.

df = pd.DataFrame({'Metros': metros, 'Niños': niños})
# Mediana
print (df.median())

# Media Aritmetica
print(df.mean()) 

# Moda
print(df.mode())

# Cuartiles
print(df.quantile([0, 0.25, 0.5, 0.75]))
```

```python
#c. Varianza y desviación Estándar.

df = pd.DataFrame({'Metros': metros, 'Niños': niños})
# Varianza
print(df.var())

# Desviacion Estandar
print(df.std())


```

```python
#d. ¿Entre qué dos valores se encuentra, como mínimo, el 75% de las observaciones?

df = pd.DataFrame({'Metros': metros, 'Niños': niños})
print(df.quantile([0, 0.75])) 

```

```python
# ¿Qué es la estadística descriptiva?
# La estadística descriptiva es la rama de la estadística que se encarga de describir, analizar y representar una serie de datos, para poder interpretarlos y resumirlos de la manera más sencilla posible.

# ¿Qué son las tablas de frecuencia?
# Una tabla de frecuencias es una tabla que resume los datos de un conjunto de observaciones, indicando el número de observaciones que se encuentran en cada una de las clases en que se divide la variable estadística.

# ¿Cuáles son las medidas de tendencia central?
# Las medidas de tendencia central son medidas de posición que nos indican el valor del centro de la distribución de los datos. Las medidas de tendencia central más utilizadas son la media, la mediana y la moda.

# ¿Qué son las medidas de dispersión?
# Las medidas de dispersión son medidas de posición que nos indican el grado de variabilidad de los datos. Las medidas de dispersión más utilizadas son el rango, la varianza y la desviación típica.
```
