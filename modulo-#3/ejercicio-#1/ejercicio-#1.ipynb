```python
#-------------------------------------------------------
# Name: Luis Reyes Sandoval
# Date: 24/04/2023
# Description: Trabajo #1
# Version: 1.0
#-------------------------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


metros = [1,2,3,4,5,6,7,8]
nios = [2,6,10,5,10,3,2,2]

#a) Tabla de Frecuencia. Diagrama de Barras para las frecuencias absolutas, relativas y acumuladas.

df = pd.DataFrame({'Metros': metros, 'Nios': nios})
df['Frecuencia Absoluta'] = df.Nios.value_counts()
df['Frecuencia Relativa'] = df.Nios.value_counts(normalize=True)
df['Frecuencia Acumulada'] = df['Frecuencia Relativa'].cumsum()
df

# grafico de barras para frecuencia absoluta
df['Frecuencia Absoluta'].plot(kind='bar', color='blue', title='Frecuencia Absoluta', xlabel='Metros', ylabel='Nios')
plt.show()
df['Frecuencia Relativa'].plot(kind='bar', color='blue', title='Frecuencia Relativa', xlabel='Metros', ylabel='Nios')
plt.show()
df['Frecuencia Acumulada'].plot(kind='bar', color='blue', title='Frecuencia Acumulada', xlabel='Metros', ylabel='Nios')
plt.show()

```

```python
#b. Mediana, media aritmtica, moda y cuartiles.

df = pd.DataFrame({'Metros': metros, 'Nios': nios})
# Mediana
print (df.median())

# Media Aritmetica
print(df.mean()) 

# Moda
print(df.mode())

# Cuartiles
print(df.quantile([0, 0.25, 0.5, 0.75]))
```

```python
#c. Varianza y desviacin Estndar.

df = pd.DataFrame({'Metros': metros, 'Nios': nios})
# Varianza
print(df.var())

# Desviacion Estandar
print(df.std())


```

```python
#d. Entre qu dos valores se encuentra, como mnimo, el 75% de las observaciones?

df = pd.DataFrame({'Metros': metros, 'Nios': nios})
print(df.quantile([0, 0.75])) 

```
