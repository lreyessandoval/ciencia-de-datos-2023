```python
#A partir del fichero del titanic.csv, realizar los siguiente:
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings

# leer csv
df = pd.read_csv('titanic.csv')
```

```python
# 1. Confeccione un gráfico  de barras  para visualizar hombres y mujeres que han sobrevivido.
# 2. Cambiar el color del gráfico.

survived = df[df['Survived'] == 1]
sns.countplot(
    data=survived,
    x='Sex',
    palette=['blue','red'],
)
```

```python
# 3. Gráfico de barras de los sobrevivientes agrupados por Pclass, Embarked mostrando el conteo x Embarked

survived = df[df['Survived'] == 1]
sns.countplot(
    data=survived,
    x='Embarked',
    hue='Pclass',
    palette=['blue','red','green'],
)

```
