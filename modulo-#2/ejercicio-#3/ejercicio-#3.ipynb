```python
#1. Crear un programa que sea capaz de leer el archivo videojuegos1.csv, 
# luego sacar el promedio de los precios y el promedio de la calificación.



import pandas as pd
import numpy as np

df = pd.read_csv('videojuegos.csv')

#dividir la columna 0 del dataframe en columnas por el separador ","
df[['Nombre', 'Calificación', 'Precio']] = df['Nombre,Calificación,Precio'].str.split(',', expand=True)
#eliminar la columna 0 del dataframe
df.drop('Nombre,Calificación,Precio', axis=1, inplace=True)
#convertir la columna Calificación en float
df['Calificación'] = df['Calificación'].astype(float)
#convertir la columna Precio en float
df['Precio'] = df['Precio'].astype(float)
#calcular el promedio de la columna Calificación
promedio_calificacion = df['Calificación'].mean()
#calcular el promedio de la columna Precio
promedio_precio = df['Precio'].mean()
#imprimir el promedio de la columna Calificación
print('El promedio de la calificación es: ', promedio_calificacion)
#imprimir el promedio de la columna Precio
print('El promedio del precio es: ', promedio_precio)
#imprimir el dataframe
df.head()

```

```python
#2.- Se tiene el archivo videojuegos2.csv con información sobre todo el contenido de su biblioteca de videojuegos.
#Note que un juego aparecerá listado en el archivo una vez por cada consola
# en la que fue lanzado.
# La categoría del juego se especifica siempre en el primer Tag. Por ejemplo,
# Double Dragon pertenece a la categoría Beat-‘em up.

#a) La función juegosConsolas(nomArchivo, categoria, decada) que recibe el
# nombre del archivo con la información de los videojuegos, una categoría y
# número de cuatro dígitos representando una década de años. La función
# retorna una tupla con 2 elementos. El primer elemento es la lista con los
# valores únicos de todos los juegos de esa década para esa categoría. El
# segundo elemento es la lista con valores únicos de todas las consolas que
# tienen juegos para esa década y categoría.



import pandas as pd
import numpy as np

##################################################################
# NOTA IMPORTANTE: El archivo fue normalizado...
##################################################################
#carga archivo y normaliza columnas...
def carga_archivo(nomArchivo):
    df = pd.read_csv('videojuegos2.csv', delimiter=';')
    
    df['year'] = df['year'].astype(int)

    return df


def juegosConsolas(nomArchivo, categoria, decada):
    df = carga_archivo(nomArchivo) 

    #convertir año a decada de 2 digitos
    decada = decada - (decada % 10)
    if (decada >= 2000):
        decada = decada - 2000
    else :
        decada = decada - 1900

    # El primer elemento es la lista con los valores únicos de todos los juegos de esa década para esa categoría. 
    df_filtro = df[(df['year'] >= decada) & (df['year'] < decada + 10) & (df['tag1'] == categoria)]

    list_juego = df_filtro['Nombre'].unique().tolist()

    # El segundo elemento es la lista con valores únicos de todas las consolas que tienen juegos para esa década y categoría.
    list_consola = df_filtro['Consola'].unique().tolist()

    #retorna una tupla con 2 elementos.
    return (list_juego, list_consola)


    

#print (carga_archivo('videojuegos2.csv').head(10))
print(juegosConsolas('videojuegos2.csv', 'RPG', 1980))

```

```python
def crearMatriz(nomArchivo, categoria, decada):
    df = carga_archivo(nomArchivo) 

    #convertir año a decada de 2 digitos
    decada = decada - (decada % 10)
    if (decada >= 2000):
        decada = decada - 2000
    else :
        decada = decada - 1900

    # El primer elemento es la lista con los valores únicos de todos los juegos de esa década para esa categoría. 
    df_filtro = df[(df['year'] >= decada) & (df['year'] < decada + 10) & (df['tag1'] == categoria)]

    list_juego = df_filtro['Nombre'].unique().tolist()

    # El segundo elemento es la lista con valores únicos de todas las consolas que tienen juegos para esa década y categoría.
    list_consola = df_filtro['Consola'].unique().tolist()

    matriz= pd.DataFrame(0, index=list_juego, columns=list_consola)

    for index, row in df_filtro.iterrows():
        matriz.loc[row['Nombre'], row['Consola']] = df[(df['Nombre'] == row['Nombre']) & (df['Consola'] == row['Consola'])]['Calificacion'].mean()


    return matriz

print("Datos:")
datos = crearMatriz('videojuegos2.csv', 'RPG', 1980)
print(datos)
print()
print("Matriz:")
print(np.matrix(datos))
```
